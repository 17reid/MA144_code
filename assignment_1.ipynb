Importing libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interact
```

Creating parametric function

```python
def P(theta,r):
    return ((1-r)*np.cos(theta)+r*np.cos(((1-r)*theta)/(r)), (1-r)*np.sin(theta) - r*np.sin(((1-r)*theta)/(r)))
```

Now make a function to plot the parametric graphs

```python
def plotP(r, rName):
    _, ax = plt.subplots()
    theta_vals = np.linspace(0,10*np.pi,3000)
    coordinates = P(theta_vals,r)
    ax.plot(coordinates[0],coordinates[1])
    ax.axis('scaled')
    ax.set_xlim(-1.5,1.5)
    ax.set_ylim(-1.5,1.5)
    ax.set_title(f"P(theta) for r = {rName}")
    
```

Make graphs

```python
plotP(1/2, "1/2")
plotP(1/4, "1/4")
plotP(2/3, "2/3")
plotP(1/np.sqrt(2), "1/sqrt(2)")
```
