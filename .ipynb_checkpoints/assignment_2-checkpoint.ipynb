Importing libraries

```python
%matplotlib ipympl
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

Defining $f(x,y)$

```python
def f(x,y):
    return (np.exp(x) + y**3 - 2*x - 3*y)
```

Getting points and plotting

```python
Nsteps = 5000

X = np.linspace(-2,2,Nsteps+1)
Y = np.linspace(-2,2,Nsteps+1)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


surf = ax.plot_surface(X,Y,Z, cmap  = cm.coolwarm, linewidth = 0, antialiased=False, alpha=0.7)
ax.scatter([np.log(2),np.log(2)],[1,-1],[f(np.log(2),1),f(np.log(2),-1)],alpha=1)
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.view_init(azim=150)
plt.show()
```
