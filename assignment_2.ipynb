Importing libraries

```python

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

Defining $f(x,y)$

```python
def f(x,y):
    return (np.exp(x) + y**3 - 2*x - 3*y)
```

Getting points and plotting

```python
Nsteps = 15000

X = np.linspace(-3,3,Nsteps+1)
Y = np.linspace(-3,3,Nsteps+1)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


surf = ax.plot_surface(X,Y,Z, cmap  = "Spectral", linewidth = 0, antialiased=False, alpha=0.5)
ax.scatter([np.log(2),np.log(2)],[1,-1],[f(np.log(2),1),f(np.log(2),-1)],s=40,marker="x",color="k",alpha=1)
ax.text(np.log(2),1,f(np.log(2),1), "(x,y) = (log(2),1)")
ax.text(np.log(2),-1,f(np.log(2),-1), "(x,y) = (log(2),-1)")
ax.text(0,2,f(0,2), "P")
ax.set_xlim(-2.5,2.5)
ax.set_ylim(-2.5,2.5)
ax.view_init(azim=160)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title("z = f(x,y)")

plt.show()
```
